# Creación de nombres de las columnas del df

```python
# Lista de columnas para mantener el orden
columns = ['Authors','Author(s) ID','Title','Year','Source title','Volume','Issue','Art. No.',
           'Page start','Page end','Page count','Cited by','DOI','Link','Affiliations','Authors with affiliations',
           'Abstract','Author Keywords','Index Keywords','ISSN','ISBN','CODEN','PubMed ID','Document Type',
           'Publication Stage','Access Type','Source','EID']
```

# Cargue de todos los archivos csv de scopus y creación de un archivo consolidado

```python
all_files = glob.glob(path + "/*.csv")

li = []

for filename in all_files:
    with open(filename, 'rb') as f:
        print(filename)
        result = chardet.detect(f.read())
        df_aux = pd.read_csv(filename, encoding=result['encoding']) # Por las letrás en portugues hay mucho problema para leerlas
    li.append(df_aux)

frame = pd.concat(li,ignore_index=True,sort=True)
frame = frame[columns]
```

```python
#Elimina los renglones duplicados
frame_unique = frame.drop_duplicates(keep=False)
```

```python
frame_unique.to_csv("scopus_bd.csv",columns=columns,sep=',',index=False)
```
